```python
z = fc_collection.design_matrix("FC", "shock", "Region1G", "Region0R")
```

```python
import pymc
design_matrix = z.copy().drop(["R", "Mouse_ID"], axis=1)
with pymc.Model() as model:
    # Define priors
    sigma = pymc.HalfCauchy("sigma", beta=10)
    intercept = pymc.Normal("Intercept", 0, sigma=20)
    slope = pymc.Normal("slope", 0, sigma=20, shape=design_matrix.shape[1])

    # Define likelihood
    likelihood = pymc.Normal("y", mu=intercept + pymc.math.dot(design_matrix, slope), sigma=sigma, observed=z["G"])

    # Inference!
    # draw 3000 posterior samples using NUTS sampling
    idata = pymc.sample(3000, cores=24)
```

```python
import arviz as az
import xarray as xr
```

```python
idata.posterior["y_model"] = idata.posterior["Intercept"] + idata.posterior[z[["X", "Y", "X_velocity", "Y_velocity", "X_acceleration", "Y_acceleration", "Freezing"]]] * xr.DataArray(z[["X", "Y", "X_velocity", "Y_velocity", "X_acceleration", "Y_acceleration", "Freezing"]])

_, ax = plt.subplots(figsize=(7, 7))
az.plot_lm(idata=idata, y="y", num_samples=100, axes=ax, y_model="y_model")
```
